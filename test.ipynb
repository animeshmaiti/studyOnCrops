```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'year': [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009],
    'RICE PRODUCTION (1000 tons)': [4000, 4200, 4300, 4500, 4600, 4800, 5000, 5200, 5400, 5600]
}
df = pd.DataFrame(data)

# Calculate the moving average
df['Moving Average'] = df['RICE PRODUCTION (1000 tons)'].rolling(window=3).mean()

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['RICE PRODUCTION (1000 tons)'], label='Annual Production', marker='o')
plt.plot(df['year'], df['Moving Average'], label='Moving Average', linestyle='--',marker='o')
plt.xlabel('Year')
plt.ylabel('Rice Production (1000 tons)')
plt.title('Rice Production Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=100)
print(data)

# Create the box plot
plt.boxplot(data, patch_artist=True)
plt.xlabel("Values")
plt.title("Box Plot Example")
plt.show()
```

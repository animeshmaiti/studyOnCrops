```python
plt.figure(figsize=(12, 6))

# Group data by year and sum the rice production
rice_production_by_year =df.groupby('Year')['RICE PRODUCTION (1000 tons)'].sum()

sns.lineplot(x=rice_production_by_year.index, y=rice_production_by_year.values, marker='o', color='skyblue')

plt.title('Rice Production (1000 tons) Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Rice Production (1000 tons)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rice_production_by_year.index, rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='State Name', y='RICE PRODUCTION (1000 tons)', 
            data=df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum().reset_index().sort_values(by='RICE PRODUCTION (1000 tons)', ascending=False),
            palette='viridis',
            hue='State Name',
            dodge=False)
plt.title('Rice Production (1000 tons) by State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Rice Production (1000 tons)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
sns.set_style("whitegrid")

for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], bins=20, color='skyblue', edgecolor='black', kde=True)
    plt.title(f'Histogram of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[numeric_columns])
plt.title('Boxplot of Numeric Columns', fontsize=14)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
